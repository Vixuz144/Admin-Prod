```python
from random import gauss, seed
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mL = 24970
dL = 3000
seed(46)

x = [27512,	25034,	27947,	19389,	29445,	23729,	28078,	39942,	19515,	26038,	33726,	20003]


l = []
for i in range(12):
  r = round(gauss(mL,dL),2)
  l.append(r)
l.extend(x)
plt.plot(l)

plt.show()
datos = pd.DataFrame([l])

print(datos)
datos.to_excel('primeros litros.xlsx')
```

```python
x = [1478,	1874,	2616,	1802, 2197,	2087,	1834,	2096,	1757,	1932,	1548,	1819]
mB = np.mean(x)
dB = np.std(x)
seed(19)

print(mB,dB)

l = []
for i in range(12):
  r = round(gauss(mB-300+i*20,dB-250+i*10))
  l.append(r)
l.extend(x)
plt.plot(l)

plt.show()
datos = pd.DataFrame([l])

print(datos)
datos.to_excel('primeros cajas.xlsx')
```
